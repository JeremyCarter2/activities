XML

3. Here is the url to an example xml file: https://www.w3schools.com/xml/cd_catalog.xml Use the requests library to download the page and parse the xml file.

```python
# Your Code Here (imports done for you)
import xml.etree.ElementTree as ET
import urllib.request, urllib.parse, urllib.error

url = 'https://www.w3schools.com/xml/cd_catalog.xml'
response = urllib.request.urlopen(url).read()

tree = ET.fromstring(response)

artists = []
titles = []
countries = []
companies = []
prices = []
years = []

for cd in tree.findall('CD'):
    title = cd.find('TITLE').text
    artist = cd.find('ARTIST').text
    country = cd.find('COUNTRY').text
    company = cd.find('COMPANY').text
    price = cd.find('PRICE').text
    year = cd.find('YEAR').text
    artists.append(artist)
    titles.append(title)
    countries.append(country)
    companies.append(company)
    prices.append(price)
    years.append(year)



```

4. Create a dataframe from the xml file containing the following columns: title, artist, country, company, price, year

```python
# Your Code Here (imports done for you)
import pandas as pd

cd_df = pd.DataFrame({'Artist': artists, 
                      'Title': titles,
                      'Country': countries,
                      'Company': companies,
                      'Price': prices,
                      'Year': years
                      })

cd_df
                
```
